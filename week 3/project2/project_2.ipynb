```python
from math import sqrt

print("Welcome to the (quadratic | cubic | quartic) roots calculator")
print("What type of equation would you like to find roots for?")
print("1. Quadratic\n2. Cubic\n3. Quartic")

choice = input("Enter your choice (1/2/3): ")
roots = []

if choice == '1':
    a = int(input("Enter the coefficient of x^2: "))
    b = int(input("Enter the coefficient of x: "))
    c = int(input("Enter the value of the constant: "))
    d = (b**2) - (4*a*c)
    
    if d < 0:
        print("\nSorry, we can't solve for complex roots!")
    elif d == 0:
        x = (-b) / (2*a)
        print(f"\nRoots: {x} twice.")
    else:
        x_1 = ((-b) + sqrt(d)) / (2*a)
        x_2 = ((-b) - sqrt(d)) / (2*a)
        print(f"\nRoots: {x_1} and {x_2}")
        
elif choice == '2':
    print("\n--- This calculator only checks for whole number roots between -100 and 100---")
    
    a = int(input("Enter the coefficient of x^3: "))
    b = int(input("Enter the coefficient of x^2: "))
    c = int(input("Enter the coefficient of x: "))
    d = int(input("Enter the value of the constant: "))
    
    for i in range(-100, 100):
        rem = a*(i**3) + b*(i**2) + c*i + d
        if rem == 0:
            roots.append(i)
        else:
            continue
        
    print(f"\nWhole number roots:")
    
    if roots:
        for root in roots:
            print(root)
    else:
        print("None")
    
elif choice == '3':
    print("\n--- This calculator only checks for whole number roots between -100 and 100---")
    
    a = int(input("Enter the coefficient of x^4: "))
    b = int(input("Enter the coefficient of x^3: "))
    c = int(input("Enter the coefficient of x^2: "))
    d = int(input("Enter the coefficient of x: "))
    e = int(input("Enter the value of the constant: "))
    
    for j in range(-100, 100):
        rem = a*(j**4) + b*(j**3) + c*(j**2) + d*j + e
        if rem == 0:
            roots.append(j)
        else:
            continue
    
    print(f"\nWhole number roots:")
    
    if roots:
        for r in roots:
            print(r)
    else:
        print("None")
        
else:
    print("Invalid choice!")
```
